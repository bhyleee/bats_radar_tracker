```python
def aggregate_date(data_directory, date):
    
    # Create list of files you will need to aggregate within directory (temp)
    merge_dir=[]
    for filename in os.listdir(data_directory): 
        if filename.endswith('.tif') and filename.startswith('classified'):
            merge_dir.append(filename)
            
        elif filename.startswith('2018'):
            merge_dir.append(filename)

        else:
            pass
            
    # Read metadata of first file
    merge_dir.sort()
    with rasterio.open(merge_dir[0]) as src0:
        meta = src0.meta
        arr1 = src0.read()
        print(arr1.shape)

    # Create aggregate file
    date_file = str(date) + '_new_aggregate.tif'
    with rasterio.open(date_file, 'w', **meta) as dst:
        dst.write(arr1)

    for id, layer in enumerate(merge_dir, start=1):
        # try:
        if layer.endswith('.tif'):
            print('layer is ' + layer)
            # print('id is ' + id)
            with rasterio.open(layer) as src1:
                array1 = src1.read()
                print(array1.shape)
                # print('src 1 is ' + src1)
                # with rasterio.open('test.tif', 'w', **meta) as dst:
                array = rasterio.open(date_file)
                print('aggregate file is ' + date_file)
                array0 = array.read(1)
                # array0 = np.where(array0 > 0.1, 1, 0)


                # array2 = np.where(array1 > 0.1, 1, 0)
                array1[np.isnan(array1)] = 0

                new_array = np.add(array0, array1)
                print('added files')
                array.close()


                with rasterio.open(date_file, 'w', **meta) as dst:
                    dst.write(new_array)
        else:
            pass


#         except:
#             pass
    # with rasterio.open('test.tif', 'w', **meta) as dst:
    #     dst.write(new_array)
```
